```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
%matplotlib inline
```

```python
df = pd.read_csv('ESSdata_Thinkful.csv')
```

```python
df.head()
```

Did people become less trusting from 2012 to 2014? Compute results for each country in the sample.

```python
plt.figure(figsize=(15, 9))
plt.ylim(4,7)
df.groupby(['cntry','year']).mean()['ppltrst'].plot(kind='bar')
```

Did people become happier from 2012 to 2014? Compute results for each country in the sample.

```python
plt.figure(figsize=(15, 9))
plt.ylim(6.5,8.5)
df.groupby(['cntry','year']).mean()['happy'].plot(kind='bar')
```

Who reported watching more TV in 2012, men or women?

```python
plt.figure(figsize=(15, 9))
plt.ylim(8300,8450)
gender_tv=df.groupby(['gndr','year']).sum().reset_index()
gender_tv[gender_tv['year']==6]['tvtot'].plot(kind='bar')
```

Who was more likely to believe people were fair in 2012, people living with a partner or people living alone?

```python
plt.figure(figsize=(15, 9))
partner_trust=df.groupby(['partner','year']).sum().reset_index()
partner_trust[partner_trust['year']==6]['ppltrst'].plot(kind='bar')
```

Pick three or four of the countries in the sample and compare how often people met socially in 2014. Are there differences, and if so, which countries stand out?

```python
SE_sclmeet = df[(df['year']==7) & (df['cntry']=='SE')]['sclmeet'].dropna()
NO_sclmeet = df[(df['year']==7) & (df['cntry']=='NO')]['sclmeet'].dropna()
CH_sclmeet = df[(df['year']==7) & (df['cntry']=='CH')]['sclmeet'].dropna()
```

```python
print(stats.ttest_ind(SE_sclmeet, NO_sclmeet))
print(stats.ttest_ind(SE_sclmeet, CH_sclmeet))
print(stats.ttest_ind(CH_sclmeet, NO_sclmeet))
```

Pick three or four of the countries in the sample and compare how often people took part in social activities, relative to others their age, in 2014. Are there differences, and if so, which countries stand out?

```python
SE_sclact = df[(df['year']==7) & (df['cntry']=='SE')]['sclact'].dropna()
NO_sclact = df[(df['year']==7) & (df['cntry']=='NO')]['sclact'].dropna()
CH_sclact = df[(df['year']==7) & (df['cntry']=='CH')]['sclact'].dropna()
```

```python
print(stats.ttest_ind(SE_sclact, NO_sclact))
print(stats.ttest_ind(SE_sclact, CH_sclact))
print(stats.ttest_ind(CH_sclact, NO_sclact))
```
